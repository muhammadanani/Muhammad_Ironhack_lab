<b>show the average CO2_Emission_Grams/Km  by Brand

```python
### Shows average emissions by brand
cars_emissions = cars.groupby("Make")['CO2 Emission Grams/Mile'].apply(lambda x: x.mean()).sort_values(ascending=False)
cars_emissions
```

<b>show the average CO2_Emission_Grams/Km  by Brand ... sorted

```python
### sorts emissions by brand in ascending order
cars_emissions = cars_emissions.sort_values(ascending = True)
cars_emissions
```

Use `pd.cut` or `pd.qcut` to create 4 groups (bins) of cars, by Year. We want to explore how cars have evolved decade by decade.

```python
cars['Year'].describe()
```

```python
## cut the data us

cars.groupby('Year')
bins = [1980, 1989, 1999, 2009, 2019]
labels = ['80s', '90s', '00s', '10s']
cars['Time Period'] = pd.cut(cars.Year, 
                            bins=bins, 
                            labels=labels, 
                            right=True)
cars
```

### Did cars consume more gas in the eighties?

show the average City_Km/Liter by year_range

```python
### sort by new time period column
cars_timeperiod = cars.groupby("Time Period")['City MPG'].apply(lambda x: x.mean()).sort_values(ascending = True)
cars_timeperiod
print ('No cars the highest ')
```

Does the drivetrain affect fuel consumption?

Do cars with automatic transmission consume more fuel than cars with manual transmission?

```python
## Your Code here
def rename_transmission(x):
    if(x.startswith('A')):
        return 'Automatic'
    else:
        return 'Manual'
```

```python
cars['Transmission'] = list(map(rename_transmission, cars['Transmission']))
```

```python
cars
```

```python
cars.groupby("Transmission")['Highway MPG', 'City MPG'].apply(lambda x: x.mean())
```

Use `groupby` and `aggregate` with different aggregation measures for different columns:

aggregate with average City_Km/Liter and the count of the Trans

```python
## groups by city mpg then aggregates with two operations on two columns
cars.groupby("Transmission").aggregate({"City MPG":"mean","Transmission":"count"})
```

aggregate with average City_Km/Liter and the minimum of the Trans

```python
cars.groupby("Transmission").aggregate({"City MPG":"min"})
```
